#### Data Exploration

Before we start creating models, let's do a little bit of data analysis and get a feeling for the shapes, distributions and relationships of our data.

1. Print `min`, `max` and `average` values for all of the features.
2. Print `covariance` tables for `age`, `ear.length` and `head.circumference`.
3. Plot `age`, `ear.length` and `head.circumference` versus the $1$ *feature* that is most correlated to each of them.

Don't forget to *encode* and *normalize* the data.

```python
# Work on Data Exploration here

### Encode non-numerical features

#ansur_df.dtypes  
ansur_df = pd.json_normalize(ansur_data)                 
ansur_df["gender"] = ansur_df["gender"].map({"M": 1, "F": 0}) 

#ansur_df.head(2)

## 1. Print min, max, avg

ansur_df.min()
ansur_df.max()
ansur_df.mean()

### Normalize all data

scaler = StandardScaler()
ansur_scaler = StandardScaler().set_output(transform="pandas")  
ansur_scaled_df = ansur_scaler.fit_transform(ansur_df)


## 2. Print Covariances

ansur_df.cov()

## 3. Plot features most correlated to age, ear length and head circumference

correlations = ansur_df.corr()

age_correlations = correlations["age"].sort_values(ascending=False)
print("age correlation:\n", age_correlations)

earlength_correlations = correlations["ear.length"].sort_values(ascending=False)
print("earlength correlations:\n", earlength_correlations)

headcirc_correlations = correlations["head.circumference"].sort_values(ascending=False)
print("hearcirc correlation:\n", headcirc_correlations)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

age_correlations.drop("age").plot(kind="bar", ax=axes[0], title="age")
earlength_correlations.drop("ear.length").plot(kind="bar", ax=axes[1], title="ear length")
headcirc_correlations.drop("head.circumference").plot(kind="bar", ax=axes[2], title="head circ")

plt.tight_layout()
plt.show()
```

### Regression

Now, we want to create a regression model to predict `head.circumference` from the data.

From our [Week 06](https://github.com/PSAM-5020-2026S-A/WK06) notebook, we can create a regression model by following these steps:

1. Load dataset (done! 🎉)
2. Encode label features as numbers (done! ⚡️)
3. Normalize the data (done! 🍾)
4. Separate the outcome variable and the input features
5. Create a regression model using all features
6. Run model on training data and measure error
7. Plot predictions and interpret results
8. Run model on test data, measure error, plot predictions, interpret results

```python
# Work on Regression Model here

## Separate outcome variable and input features

y = ansur_scaled_df["age"]
x = ansur_scaled_df.drop(columns=["age"])    

## Create a regression model

ansur_model = LinearRegression()
ansur_model.fit(x,y)

## Measure error on training data

predicted_age = ansur_model.predict(x)
error = regression_error(y, predicted_age)
print(f"reg error: {error:.2f}")

## Plot predictions and interpret results

plt.scatter(ansur_df["age"], ansur_df["ear.length"])
plt.xlabel("age")
plt.ylabel("ear.length")
plt.show() 

plt.scatter(predicted_age, ansur_df["ear.length"], color='r', marker='o', alpha=0.05, s=3)
plt.xlabel("age")
plt.ylabel("ear.length")
plt.show() 

plt.scatter(ansur_df["age"], ansur_df["ear.length"],color='b', marker='o', alpha=0.3, s=3, label="Actual")
plt.scatter(predicted_age, ansur_df["ear.length"],color='r', marker='o', alpha=0.3, s=3, label="Predicted")
plt.xlabel("age")
plt.ylabel("ear.length")
plt.legend()
plt.show()
```

```python
## Run model on test data
y_test = ansur_test_scaled_df["age"]
x_test = ansur_test_scaled_df.drop(columns=["age"])

predicted_test_age = ansur_model.predict(x_test) 

## Measure error on test data
test_error = regression_error(y_test, predicted_test_age)
train_error = regression_error(y, predicted_age)

print(f"Train error: {train_error:.2f}")
print(f"Test error:  {test_error:.2f}")

## Plot predictions and interpret results

plt.scatter(ansur_test_df["age"], ansur_test_df["ear.length"],
            color='b', marker='o', alpha=0.3, s=3, label="Actual")
plt.scatter(predicted_test_age, ansur_test_df["ear.length"],
            color='r', marker='o', alpha=0.3, s=3, label="Predicted")
plt.xlabel("age")
plt.ylabel("ear.length")
plt.title("Test Data — Actual vs Predicted")
plt.legend()
plt.show()
```
